```python
import requests
from bs4 import BeautifulSoup

page = requests.get("http://smartlab.tmit.bme.hu/oktatas-deep-learning")   #obtaining and prettifying the HTML code
soup = BeautifulSoup(page.text, 'html.parser')
soup.prettify

texts = soup.get_text()
print(texts)

# you can also apply the 'replace' method to get rid of the newlines and tabs (see next cell), 
# but I think it's nicer to print out this way
```

```python
# Histogram

# the letter counters
import string
d = dict.fromkeys(string.ascii_lowercase, 0)
for letter in ['', '', '', '', '', '', '', '', '']:
    d[letter] = 0

# we are not going to take multiple digit letters ('ny', 'dzs' etc.) into consideration, altough there exists a
# Hungarian tokenzier called 'quntoken' which can handle those letters, but I couldn't get it to work on my computer


# sorting the keys to match the Hungarian alphabet
abc = 'abcdefghijklmnopqrstuvwxyz'
d = dict(sorted(d.items(), key=lambda i: abc.index(i[0])))



# formatting the texts and counting the letters
texts = texts.lower().replace(u'\n','').replace(u'\t', '').replace(u'\ufeff', '').replace(" ","")
for letter in texts:
    if letter in d.keys():
        d[letter] += 1

#plotting the histogram
plt.figure(figsize=(10,6))
plt.title("Distribution of letters")
plt.grid(axis='y',zorder=0)
plt.bar(list(d.keys()), np.asarray(list(d.values()))/sum(d.values()), zorder=3)
plt.show()
```
